```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Generate example signal (sum of two sine waves)
time = np.linspace(0, 1, 1000, endpoint=False)
s1 = np.sin(2 * np.pi * 25 * time)  # 25 Hz sine wave
s2 = np.sin(2 * np.pi * 50 * time)  # 50 Hz sine wave
signal = s1 + s2

# Add noise to the signal
noise_level = 0.5
signal += noise_level * np.random.randn(len(time))

# Perform Continuous Wavelet Transform (CWT)
wavelet = 'morl'  # Morlet wavelet
scales = np.arange(1, 101)  # Range of scales (frequencies)
cwt_result, frequencies = pywt.cwt(signal, scales, wavelet, sampling_period=1)

# Plot the original signal
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(time, signal)
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# Plot the CWT result
plt.subplot(2, 1, 2)
plt.imshow(np.abs(cwt_result), extent=[0, 1, frequencies[-1], frequencies[0]], aspect='auto', cmap='jet')
plt.title('Continuous Wavelet Transform')
plt.xlabel('Time')
plt.ylabel('Frequency (Hz)')
plt.colorbar(label='Magnitude')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# List of commonly used wavelets
wavelet_names = pywt.wavelist()

# Sampling points for the x-axis
x = np.linspace(-2, 2, num=1000)

# Plot each wavelet
for wavelet_name in wavelet_names:
    wavelet = pywt.Wavelet(wavelet_name)
    psi, _ = wavelet.wavefun()

    plt.figure()
    plt.title(wavelet_name)
    plt.plot(x, psi)
    plt.xlabel('x')
    plt.ylabel('ψ(x)')
    plt.grid()

plt.show()
```
